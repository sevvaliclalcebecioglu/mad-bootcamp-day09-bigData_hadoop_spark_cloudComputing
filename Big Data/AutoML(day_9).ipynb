#### *Regression*

```python
import pandas as pd
```

```python
from pycaret.regression import *
```

```python
df=pd.read_pickle('kc_house.pkl')
```

```python
setup(df,target='price',session_id=123)

# bütün regression işlemlerini tek satırda yapmış olduk
```

```python
best_model=compare_models()

# 20 tane regression modeli deniyor, küçükten büyüğe listeliyor.
```

```python
plot_model(best_model,plot='error')
```

```python
plot_model(best_model,plot='learning')
```

```python
plot_model(best_model,plot='manifold')
```

```python
plot_model(best_model,plot='feature')
```

```python
#plot_model(best_model,plot='rfe')

# recursive feature elimination
```

```python
plot_model(best_model,plot='feature_all')
```

#### *Classification*

```python
df=pd.read_csv('pima-indians-diabetes.csv')
```

```python
from pycaret.classification import *
```

```python
setup(df,target='Outcome')
```

```python
best_model=compare_models()
```

```python
plot_model(best_model,plot='auc')
```

```python
plot_model(best_model,plot='feature')
```

```python
plot_model(best_model,plot='confusion_matrix')
```

```python
plot_model(best_model,plot='learning')
```

```python
plot_model(best_model,plot='manifold')
```

```python
plot_model(best_model,plot='dimension')
```

```python
save_model(best_model,'best_classifier')

# oluşturduğum modeli kaydetmek için;
```

```python
model=load_model('best_classifier')

# kaydettiğim modeli kullanmak için;
```

```python
plot_model(model,plot='class_report')
```

```python
plot_model(model,plot='boundary')
```
