```python
import pandas as pd

# Pandas kütüphanesini veri analizi ve tablo işlemleri için yükler
```

```python
df = pd.read_excel('cars.xls')

# cars.xls dosyasını pandas DataFrame olarak yükler
```

```python
#pip install summarytools
```

```python
from summarytools import dfSummary

# summarytools kütüphanesinden dfSummary fonksiyonunu içe aktarır, veri setinin özet istatistiklerini hızlıca oluşturmak için
```

```python
dfSummary(df)

# summarytools paketi ile veri setinin özet istatistiklerini (EDA) tek satırda oluşturur
```

```python
#pip install dtale
```

```python
import dtale
```

```python
d = dtale.show(df)

# dtale kütüphanesi ile DataFrame'i interaktif bir web arayüzünde görüntüler, veri keşfini kolaylaştırır
```

```python
d.open_browser()

# dtale ile DataFrame'in interaktif özetini ve analiz araçlarını tarayıcıda yeni sekmede açar
```

```python
#pip install ydata-profiling
```

```python
from ydata_profiling import ProfileReport

# ydata_profiling kütüphanesinden ProfileReport fonksiyonunu içe aktarır, veri seti için detaylı EDA raporu oluşturmak için
```

```python
p = ProfileReport(df)

# Veri seti için ydata_profiling kullanarak detaylı bir EDA (Exploratory Data Analysis) raporu hazırlar
```

```python
p.to_file('eda.html')

# Oluşturulan EDA raporunu 'eda.html' dosyası olarak kaydeder, tarayıcıda görüntüleyebiliriz
```
