# _Web Scraping - İnternetten Veri Kazıma_

```python
from bs4 import BeautifulSoup as soup

# HTML ve XML verilerini kolayca parse etmek için BeautifulSoup kütüphanesini "soup" olarak kullanıyoruz.
```

```python
import requests

# Web sayfalarından veri çekmek için HTTP istekleri göndermemizi sağlayan "requests" kütüphanesini dahil ediyoruz.
```

```python
page = requests.get('http://localhost:8080/')

# Belirtilen URL'ye HTTP GET isteği göndererek sayfanın HTML içeriğini "page" değişkenine alıyoruz.
```

```python
page.content

# HTTP GET isteği ile aldığımız sayfanın ham HTML içeriğini byte formatında gösterir.
```

```python
sp = soup(page.content, 'html.parser')

# HTML içeriğini BeautifulSoup ile parse ediyoruz
```

```python
sp2 = sp.prettify()

# Parse edilmiş HTML içeriğini daha okunabilir bir formatta string olarak alıyoruz
```

```python
sp2

# Prettify ile formatlanmış HTML içeriğini görüntülemek için değişkeni çağırıyoruz
```

```python
siir = sp.find('p')

# HTML içeriğinde ilk <p> etiketini bulup değişkene atıyoruz
```

```python
siir

# Bulunan <p> etiketini ve içeriğini görüntülemek için değişkeni çağırıyoruz
```

```python
siir = sp.find_all('p')

# HTML içeriğindeki tüm <p> etiketlerini bulup liste halinde değişkene atıyoruz
```

```python
siir[3]

# Bulunan <p> etiketleri listesinin 4. elemanını (index 3) görüntülüyoruz
```

```python
siir[0].text.replace('<br>', ' ').replace('\r', ' ').replace('\n', ' ')

# İlk <p> etiketinin içindeki metni alıp satır sonlarını ve <br> etiketlerini boşlukla değiştiriyoruz
```
